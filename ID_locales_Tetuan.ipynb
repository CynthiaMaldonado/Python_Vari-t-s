# Spot-a-Shop

### Limpieza de datos del dataset de licencias del Ayuntamiento de Madrid para filtrar, dentro del barrio de Tetuán (barrio elegido para para la presentación de la webapp) los establecimientos abiertos que cumplen las características que buscábamos.

```python
import pandas as pd
```

```python
places = pd.read_csv('OPEN DATA Locales-Epigrafes202007.csv', sep=';', encoding = "ISO-8859-1" )
places.head()
```

```python
places.info()  # Hay columnas aparentemente importantes que están vacias
```

```python
# examinamos barrios
places['desc_distrito_local'].value_counts()
```

```python
# examinamos grupos
places['desc_seccion'].value_counts()
```

```python
tetuan = places.copy()
```

```python
# tras intentos fallidos de seleccionar el barrio con boleano y no conseguirlo (tipo de codificación)?, lo consigo con el principio del str
tetuan = tetuan.loc[tetuan['desc_distrito_local'].str.startswith('T')]
```

```python
# Nos quedamos solo con las secciones uqe incluyen los tipos de establecimieto que buscamos
tetuan = tetuan.loc[tetuan['desc_seccion'].isin(['COMERCIO AL POR MAYOR Y AL POR MENOR; REPARACI0N DE VEHICULOS DE MOTOR Y MOTOCICLETAS', 'HOSTELERIA'])]
tetuan
```

```python
# Elegimos solo los que eaparecen abiertos
tetuan = tetuan[tetuan['desc_situacion_local'] == 'Abierto']
tetuan.shape
```

```python
# Nos quedamos con las columnas que necesitamos
tetuan = tetuan[['id_local', 'coordenada_x_local', 'coordenada_x_local', 'desc_vial_edificio', 'id_ndp_edificio', 'num_edificio', 'rotulo','id_epigrafe', 'desc_epigrafe']]
tetuan
```

```python
tetuan.desc_epigrafe.value_counts()
```

```python
tetuan['id_epigrafe'].value_counts()
```

```python
tetuan.loc[tetuan['id_epigrafe'] == 471103] # valorar tiendas de conveniencia
```

```python
tetuan.loc[tetuan['id_epigrafe'] == 472906] # valorar tiendas de golosinas
```

```python
# Nos quedamos con las descripciones de epígrafe que contienen palabras clave
L = ['ferre', 'pan', 'frutas', 'papel', 'platos', 'secos', 'golo']
pat = r'(\b{}\b)'.format('|'.join(L))
tetuan_final = tetuan[tetuan["desc_epigrafe"].str.contains(pat, case=False, na=False)]
```

```python
tetuan_final.desc_epigrafe.value_counts()
```

```python
tetuan_final['id_epigrafe'].value_counts() # Epígrafes del índice de la comunidad
```

```python
# Eliminamos las que sobran según nuestras preferencias con las descripciones de epígrafe que contienen estas palabras clave
N = ['metales', 'mayor']
pat = r'(\b{}\b)'.format('|'.join(N))
tetuan_final = tetuan_final[~tetuan_final["desc_epigrafe"].str.contains(pat, case=False, na=False)]
```

```python
# Selección final de tipos de negocios que se decidieron previamente 
tetuan_final.desc_epigrafe.value_counts()
```
